```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
age=[21,22,23,24,25]
salary=[50000,60000,70000,45000,40000]
bonus=[60000,40000,45000,70000,60000]
```

```python
plt.plot(age,salary)
```

```python
plt.xlim(21,26) # x -axis
plt.ylim(40000,80000) # y -axis
plt.xlabel('Age')
plt.ylabel('Salary')
plt.plot(age,salary)
#plt.style.use('default')
plt.show()
```

```python
plt.title('Salary Analysis')
plt.xticks((21,22,23,24,25),('saad','zubair','haris','anas','ali'))
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.plot(age,salary)
plt.plot(age,bonus, color='g')
plt.legend(['Base Salary','Bonus'])
#plt.style.use('classic')
plt.show()
```

```python
plt.title('Salary Analysis')
plt.xticks((21,22,23,24,25),('saad','zubair','haris','anas','ali'))
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.plot(age,salary,'4-.r',label='Base Salary')
plt.plot(age,bonus, color='g',label='Bonus')
plt.legend()
#plt.style.use('grayscale')
plt.show()
```

```python
plt.title('Salary Analysis')
plt.xticks((21,22,23,24,25),('saad','zubair','haris','anas','ali'))
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.plot(age,salary,'x--c',label='Base Salary')
plt.plot(age,bonus, color='g',linestyle=":", marker='x' ,label='Bonus')
plt.legend()
#plt.style.use('dark_background')
plt.show()
```
