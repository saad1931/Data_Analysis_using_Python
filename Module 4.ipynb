```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#plt.style.use('dark_background')
plt.style.use('default')
```

```python
age=[21,22,23,24,25]
salary=[50000,60000,70000,45000,40000]
bonus=[60000,40000,45000,70000,60000]
```

```python
plt.plot(age,salary)
```

```python
plt.xlim(21,26) # x -axis
plt.ylim(40000,80000) # y -axis
plt.xlabel('Age')
plt.ylabel('Salary')
plt.plot(age,salary)
#plt.style.use('default')
plt.show()
```

```python
plt.title('Salary Analysis')
plt.xticks((21,22,23,24,25),('saad','zubair','haris','anas','ali'))
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.plot(age,salary)
plt.plot(age,bonus, color='g')
plt.legend(['Base Salary','Bonus'])
#plt.style.use('classic')
plt.show()
```

```python
plt.title('Salary Analysis')
plt.xticks((21,22,23,24,25),('saad','zubair','haris','anas','ali'))
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.plot(age,salary,'4-.r',label='Base Salary')
plt.plot(age,bonus, color='g',label='Bonus')
plt.legend()
#plt.style.use('grayscale')
plt.show()
```

```python
plt.title('Salary Analysis')
plt.xticks((21,22,23,24,25),('saad','zubair','haris','anas','ali'))
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.plot(age,salary,'x--c',label='Base Salary')
plt.plot(age,bonus, marker='x' ,linestyle=":", color='g',label='Bonus')
plt.legend()
#plt.style.use('dark_background')
plt.show()
```

## Line plot

### iris dataset

```python
iris =sns.load_dataset('iris')
```

```python
iris.info()
```

```python
iris.head()
```

```python
iris.drop(['species'],axis=1).plot.line(title='Iris Dataset')
```

```python
columns = iris.columns.drop('species')

x_data = range(0,iris.shape[0])

fig,ax = plt.subplots()

for column in columns:
    ax.plot(x_data,iris[column],label=column)

    
ax.set_title('Iris Dataset')
ax.legend()
```

```python
sns.lineplot(data=iris.drop(['species'],axis=1))
```

## Bar plots

```python
fig,ax = plt.subplots()

ndata = iris['petal_length'].value_counts()

points = ndata.index
frequency =ndata.values

ax.bar(points,frequency)

ax.set_title('Petal Length')
ax.set_xlabel('Petal Length')
ax.set_ylabel('Frequency')
```

```python
sns.countplot(iris['petal_length'])
```

## Scatter Plots

```python
fig,ax=plt.subplots()

ax.scatter(iris['sepal_length'],iris['petal_width'])

ax.set_title('Iris Data')
ax.set_ylabel('petal width')
```

```python
colors = {'setosa': 'r','versicolor': 'g',  'virginica' : 'b'}

fig,ax=plt.subplots()

for i in range(len(iris['sepal_length'])):
    ax.scatter(iris['sepal_length'][i],iris['sepal_width'][i],color=colors[iris['species'][i]])

ax.set_title('Iris Data')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
```

```python
sns.regplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
sns.lmplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
```

```python
sns.swarmplot(x='species',y='sepal_width',data=iris)
```
